Youtube Scrape

````python
# Set up Selenium
options = Options()
options.add_argument('--headless')
options.binary_location = "/opt/firefox/firefox"
driver = webdriver.Firefox(service=FirefoxService('/usr/local/bin/geckodriver'), options=options)

# YouTube search query
keyword = "grocery"
search_url = f"https://www.youtube.com/results?search_query={keyword}"
driver.get(search_url)

# Wait for search results to load
try:
    WebDriverWait(driver, 20).until(  # Increased timeout to 20 seconds
        EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'a#video-title'))
    )
    # Collect video URLs (increase limit to 20 or more)
    video_elements = driver.find_elements(By.CSS_SELECTOR, 'a#video-title')[:20]  # Fetch the first 20 videos
    video_urls = [el.get_attribute('href') for el in video_elements]

    print("Collected URLs:")
    for url in video_urls:
        print(url)

except Exception as e:
    print(f"Error occurred while searching: {e}")
finally:
    driver.quit()

# Step 2: Use yt-dlp to Extract Metadata
ydl_opts = {
    'quiet': True,
    'skip_download': True,  # Do not download the video, just fetch metadata
}

video_metadata = []
for url in video_urls:
    try:
        with YoutubeDL(ydl_opts) as ydl:
            info = ydl.extract_info(url, download=False)
            video_metadata.append({
                "Title": info['title'],
                "Uploader": info['uploader'],
                "Duration": info['duration'],
                "Views": info['view_count'],
                "Description": info.get('description', ''),
            })
    except Exception as e:
        print(f"Error fetching metadata for {url}: {e}")

# Save metadata to a CSV file
metadata_df = pd.DataFrame(video_metadata)
csv_filename = "youtube_metadata.csv"
metadata_df.to_csv(csv_filename, index=False)

print(f"Metadata saved to {csv_filename}")
````

````python
# Load the metadata from JSON
json_filename = "youtube_metadata.json"
with open(json_filename, "r") as json_file:
    video_metadata = json.load(json_file)

print(f"Loaded {len(video_metadata)} videos.")
````

````python
# Load SpaCy English model
nlp = spacy.load("en_core_web_sm")

def preprocess_text_spacy(text):
    doc = nlp(text.lower())  # Lowercase the text
    tokens = [token.lemma_ for token in doc if not token.is_stop and not token.is_punct]
    return " ".join(tokens)

# Apply preprocessing to titles
for video in video_metadata:
    video['Processed_Title'] = preprocess_text_spacy(video['Title'])

# Display sample preprocessed data
print("Sample Processed Data:")
for video in video_metadata[:3]:
    print(f"Title: {video['Title']}")
    print(f"Processed Title: {video['Processed_Title']}")
````

````python
# Convert metadata to DataFrame
df = pd.DataFrame(video_metadata)

# Dynamically repeat and truncate the category list to match the DataFrame length
categories = ['fruits', 'meat', 'vegetables', 'dairy', 'grains']
df['Category'] = (categories * (len(df) // len(categories) + 1))[:len(df)]

print("Sample DataFrame:")
print(df[['Title', 'Processed_Title', 'Category']].head())
````

````python
# Vectorize the processed titles
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df['Processed_Title'])
y = df['Category']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train a Random Forest Classifier
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

# Evaluate the model
y_pred = model.predict(X_test)
print("Classification Report:")
print(classification_report(y_test, y_pred))
print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
````

Prediction

````python
# Predict categories for the entire dataset
df['Predicted_Category'] = model.predict(vectorizer.transform(df['Processed_Title']))

# Display predictions
print("Sample Predictions:")
print(df[['Title', 'Processed_Title', 'Category', 'Predicted_Category']].head())
````
